```python
a=int(input("enter the value of a"))
b=int(input("enter the value of b"))
print(a)
print(b)
print("addition",a+b)
print("substraction",a-b)
print("multiplication",a*b)
print("division",a/b)
print("modulous",a%b)
print("exponent",a**b)
print("floor division",a//b)
```

```python
print("assignment operators")
x=10
print("initial value of x is",x)
x+=5
print("after x+=5",x)
x-=3
print("after x-=3",x)
x*=2
print("after x*=2",x)
x/=4
print("after x/=4",x)
```

```python
print("comparison or relational operators")
a=int(input("enter a value"))
b=int(input("enter b value"))
print("a==b",a==b)
print("a!=b",a!=b)
print("a>b",a>b)
print("a<b",a<b)
print("a>=b",a>=b)
print("a<=b",a<=b)
```

```python
print("logical operators")
a=10
b=5
c=8
print("values:a",a,"values:b",b,"values:c",c)
print("a>b and a>c",a>b and a>c)
print("a>c or a>b",a>c or a>b)
print("not(a>b)",not(a>b))
print("not(a>c)",not(a>c))
```

```python
x=50
y=50
print("small integers")
print("x=",x,"y=",y)
print("x==y",x==y)
print("x is y",x is y)
print("large integers")
a=1000
b=1000
print("a=",a,"b=",b)
print("a==b",a==b)
print("a is b",a is b)
c=[1]
d=[1]
print("small integers")
print("c=",c,"d=",d)
print("c==d",c==d)
print("c is d",c is d)
```

```python
print("memebership operator")
fruits=["apple","orange","mango"]
print("fruits")
print("apple"in fruits)
print("banana"in fruits)
print("orange"not in fruits)
print("pinapple"not in fruits)
```
